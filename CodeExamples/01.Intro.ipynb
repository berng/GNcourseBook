# Random Graphs

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
N=20
P=0.3
G = nx.erdos_renyi_graph(N, P, seed=42)
plt.figure(figsize=(7, 7))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True)
```

```python
num_nodes = G.number_of_nodes()
num_edges = G.number_of_edges()
avg_degree = sum(dict(G.degree()).values()) / num_nodes if num_nodes > 0 else 0
print("Nodes: ",num_nodes," Edges: ",num_edges,"Avg Degree: ",avg_degree, 'Diameter: ', nx.diameter(G))
degrees = [d for n, d in G.degree()]
plt.hist(degrees,bins=round(np.sqrt(N))+1,density=True)
from scipy.stats import poisson
dist=poisson((N-1)*P)
l=np.array(range(N))
plt.plot(l,dist.pmf(l))
```

```python
nx.clustering(G)
```

```python
nx.average_clustering(G)
```

```python
nx.transitivity(G)
```

## Модель Барабаши — Альберт (модель предпочтительного соединения)

```python
m0=2
N=200
G = nx.barabasi_albert_graph(N, m0, seed=42)
plt.figure(figsize=(7, 7))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True)
```

```python
num_nodes = G.number_of_nodes()
num_edges = G.number_of_edges()
avg_degree = sum(dict(G.degree()).values()) / num_nodes if num_nodes > 0 else 0
print("Nodes: ",num_nodes," Edges: ",num_edges,"Avg Degree: ",avg_degree, 'Diameter: ', nx.diameter(G))
degrees = [d for n, d in G.degree()]
plt.hist(degrees,bins=round(np.sqrt(N))+1,density=True)
from scipy.stats import poisson
l=np.array(range(1,N)).astype(float)
plt.plot(l,2*(m0**2)* (l**(-3)))
plt.xscale('log')
plt.yscale('log')
```

```python
nx.clustering(G)
```

```python
nx.average_clustering(G)
```

```python
nx.transitivity(G)
```

## Модель Уоттса-Строгаца (модель тесного мира)

```python
 #   N (int): Number of nodes
 #   m0 (int): Each node is connected to k nearest neighbors (k must be even)
 #   P (float): Probability of rewiring each edge (0 <= p <= 1)
N=20
G = nx.watts_strogatz_graph(N, m0, P, seed=42)
plt.figure(figsize=(7, 7))
# pos = nx.spring_layout(G, seed=42)
pos = nx.circular_layout(G)
nx.draw(G, pos, with_labels=True)
```

```python
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True)
```

```python
pos = nx.kamada_kawai_layout(G)
nx.draw(G, pos, with_labels=True)
```

```python
pos = nx.random_layout(G, seed=42)
nx.draw(G, pos, with_labels=True)
```

```python
num_nodes = G.number_of_nodes()
num_edges = G.number_of_edges()
avg_degree = sum(dict(G.degree()).values()) / num_nodes if num_nodes > 0 else 0
print("Nodes: ",num_nodes," Edges: ",num_edges,"Avg Degree: ",avg_degree,'Diameter: ', nx.diameter(G), 'Radius: ',nx.radius(G))
print('Center:', nx.center(G))
degrees = [d for n, d in G.degree()]
plt.hist(degrees,bins=round(np.sqrt(N))+1,density=True)
```

```python
nx.clustering(G)
```
